```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import ttest_1samp

data = [
    {'Student': 'Magesh', 'Marks': 85, 'Study_Hours': 5, 'Attendance': 90},
    {'Student': 'Yoga', 'Marks': 78, 'Study_Hours': 4, 'Attendance': 85},
    {'Student': 'Uma', 'Marks': 92, 'Study_Hours': 6, 'Attendance': 95},
    {'Student': 'Karthika', 'Marks': 88, 'Study_Hours': 5, 'Attendance': 92},
    {'Student': 'Deepa', 'Marks': 80, 'Study_Hours': 4, 'Attendance': 88},
]

df = pd.DataFrame(data)

print(df)

# Summary statistics
print(df.describe())

# Correlation
print(df.corr(numeric_only=True))

# Linear Regression: Study_Hours vs Marks
X = df[['Study_Hours']]
y = df['Marks']

model = LinearRegression().fit(X, y)

# Predict marks for 7 study hours
print("Predicted Marks:", model.predict([[7]])[0])

coef = model.coef_[0]
intercept = model.intercept_

print("Coefficient:", coef)
print("Intercept:", intercept)

# Plot
plt.plot(df['Student'], df['Marks'], marker='o')
plt.title('Student Marks')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.show()
```
